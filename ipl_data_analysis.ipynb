__NOW LETS  FIND Which stadium has the highest scoring games__

```python
#lets clean itabs
df['1st_inning_score'].fillna(0,inplace=True)
df['2nd_inning_score'].fillna(0,inplace=True)
```

```python
def extval(txt):
    s=str(txt)
    data=''
    for i in s:
        if i=='/':
            break
        else:
            data+=i
    return data
```

```python
df['1st_inning_score']=df['1st_inning_score'].apply(extval).astype(int)
df['2nd_inning_score']=df['2nd_inning_score'].apply(extval).astype(int)
```

```python
df['total_runs']=df['1st_inning_score']+df['2nd_inning_score']
```

```python
df[df['total_runs']==df['total_runs'].max()][['total_runs','venue_name']]
```

__Now lets see the top batsmen__

```python
df.info()
```

```python
batsmen = pd.concat([df['home_key_batsman'], df['away_key_batsman']])

batsmen = batsmen.str.split(',')

batsmen = batsmen.explode()

batsmen = batsmen.str.strip()

```

```python
top_batsmen = batsmen.value_counts().head(10)
top_batsmen
```

```python
#we can also plot a graph
sns.barplot(top_batsmen,palette='coolwarm',gap=0.3)
```

```python
data=pd.crosstab(df['toss_won'], df['winner']).sort_values(ascending=False,by='toss_won')
```

```python
df['toss_match_win'] = df.apply(
    lambda x: 'Yes' if x['toss_won'] == x['winner'] else 'No',
    axis=1
)
```

```python
sns.countplot(data=df, x='toss_match_win')

plt.title("Did Toss Winner Also Win the Match?")
plt.show()
```
